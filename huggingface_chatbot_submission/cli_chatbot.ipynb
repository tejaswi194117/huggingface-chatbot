```python
from transformers import T5ForConditionalGeneration, T5Tokenizer

def load_model(model_name="google/flan-t5-large"):
    print(f" Loading model: {model_name}")
    tokenizer = T5Tokenizer.from_pretrained(model_name)
    model = T5ForConditionalGeneration.from_pretrained(model_name)
    print(" Model loaded!")
    return model, tokenizer
```

next is chat memory class prep
as per the task there can be 3 to 5 chat memories
so i kept it as 5


```python
from collections import deque

class ChatMemory:
    def __init__(self, max_turns=5):
        self.memory = deque(maxlen=max_turns)

    def append(self, user, bot):
        self.memory.append((user, bot))

    def get_context(self):
        return "\n".join([f"User: {u}\nBot: {b}" for u, b in self.memory])

def generate_reply(model, tokenizer, memory, user_input):
    context = memory.get_context()
    prompt = f"{context}\nAnswer the following question: {user_input}".strip()

    inputs = tokenizer(prompt, return_tensors="pt", padding=True)
    output_ids = model.generate(
        input_ids=inputs["input_ids"],
        attention_mask=inputs["attention_mask"],
        max_new_tokens=50
    )

    bot_reply = tokenizer.decode(output_ids[0], skip_special_tokens=True)
    bot_reply = bot_reply.replace("Bot: ", "").strip().capitalize()
    memory.append(user_input, bot_reply)
    return bot_reply
```

interface for the chat


```python
model, tokenizer = load_model()
memory = ChatMemory(max_turns=5)
```

```python
print(" flan-t5-large Chatbot is ready! Type '/exit' to quit.\n")

while True:
    user_input = input("User: ").strip()

    if user_input.lower() == "/exit":
        print(" Exiting chatbot. Goodbye!")
        break
    elif user_input == "":
        print(" Please type something.")
    else:
        reply = generate_reply(model, tokenizer, memory, user_input)
        print(f"Bot: {reply}\n")

```
